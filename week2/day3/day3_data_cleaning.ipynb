```python
df_clean = df.dropna(subset=["capacity"])
df_clean
```

```python
df_clean.duplicated()
```

```python
df_clean[df_clean.duplicated()]
df_clean = df_clean.drop_duplicates()
df_clean
```

```python
df_clean.isna().sum()
```

```python
mean_temp = df_clean["temperature"].mean()
mean_temp
```

```python
df_clean["temperature"] = df_clean["temperature"].fillna(mean_temp)
df_clean
```
