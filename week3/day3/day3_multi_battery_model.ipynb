```python
import pandas as pd 
import numpy as np 
np.random.seed(42)
```

```python
# Define simulation parameter
data = []
num_batteries = 20
num_cycles = 100
time_points = 50
```

```python
for battery_id in range(1, num_batteries +1):

    degradation_rate = np.random.uniform(0.003, 0.008)

    for cycle in range(1, num_cycles +1):
        for t in range(time_points):
            base_voltage = 4.2 - degradation_rate * cycle

            voltage = base_voltage - 0.02 * t + np.random.normal(0, 0.01)
            temperature = 25 + 0.03 * cycle + np.random.normal(0,0.2)

            data.append([
                battery_id, cycle,t, voltage, temperature
            ])
```
